```python
df = pd.read_excel("Canopy_Abs_Conc_Metstation_Merge.xlsx",parse_dates=['date/time'])
```

```python
df_year=df[(df['year']==2018)& (df['month'] == 6)]
```

```python
df_year.dropna()
```

```python
wd=df_year.wdir_F
conc=df_year.conc_C
ax = WindroseAxes.from_ax()
ax.contourf(wd, conc, bins=np.arange(0.5, 2.5, 0.2), cmap=cm.hot,)
ax.contour(wd, conc, bins=np.arange(0.5, 2.5, 0.2), colors='black',)
ax.set_legend()
ax.set_title('Canopy Hg Pollution Rose for Year 2018')
```

```python
wd=df_year.wdir_F
ws=df_year.wspd_F
ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
ax.set_title('Canopy Hg Pollution Rose for June 2018')
```
